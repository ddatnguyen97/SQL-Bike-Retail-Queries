```python
query = '''
        select p.product_name, b.brand_name, c.category_name, s.store_name, p.list_price, st.quantity
        from products p join brands b on p.brand_id = b.brand_id
                        join categories c on p.category_id = c.category_id
                        join stocks st on p.product_id = st.product_id
                        join stores s on st.store_id = s.store_id
        where st.quantity <= 10
        order by st.quantity desc
'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''
        select p.product_name, oi.quantity, oi.list_price, oi.discount, oi.quantity * oi.list_price * (1 - oi.discount) as revenue
        from products p join order_items oi on p.product_id = oi.product_id
'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''
        select c.category_name, p.model_year, sum(oi.quantity) as total_quantity , sum(oi.quantity * oi.list_price * oi.discount) as revenue
        from products p join order_items oi on p.product_id = oi.product_id
                        join categories c on p.category_id = c.category_id
        group by c.category_name, p.model_year
        order by c.category_name
'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''
        select subquery.category_name, max(subquery.revenue) as highest_revenue, min(subquery.revenue) as lowest_revenue
        from
        (select c.category_name, p.model_year, sum(oi.quantity) as total_quantity , sum(oi.quantity * oi.list_price * oi.discount) as revenue
        from products p join order_items oi on p.product_id = oi.product_id
                        join categories c on p.category_id = c.category_id
        group by c.category_name, p.model_year) as subquery
        group by subquery.category_name
        order by subquery.category_name
'''
df = pd.read_sql(query, connection)
df
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
query = '''
        select s.store_name, sum(oi.quantity * oi.list_price) as total_amounts
        from orders o join order_items oi on o.order_id = oi.order_id
                      join stores s on o.store_id = s.store_id        
        group by s.store_name
        order by total_amounts desc
'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''
        select c.first_name + c.last_name as full_name, o.order_id, o.order_date, sum(oi.quantity * oi.list_price) as total_amounts
        from orders o join order_items oi on o.order_id = oi.order_id
                      join customers c on o.customer_id = c.customer_id        
        group by c.first_name + c.last_name, o.order_id, o.order_date     
'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''
        select c.first_name + ' ' + c.last_name as full_name, o.order_id, o.order_date, o.shipped_date, o.required_date, abs(datediff(day, o.shipped_date, o.required_date)) as due_date
        from orders o join order_items oi on o.order_id = oi.order_id
                      join customers c on o.customer_id = c.customer_id
        where o.shipped_date > o.required_date
        group by c.first_name + ' ' + c.last_name, o.order_id, o.order_date, o.shipped_date, o.required_date
'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''
        select c.first_name + ' ' + c.last_name as full_name, count(o.order_id) as total_orders, sum(oi.quantity * oi.list_price * (1 - oi.discount)) as total_spent
        from orders o join customers c on o.customer_id = c.customer_id
                      join order_items oi on o.order_id = oi.order_id
        group by c.customer_id, c.first_name + ' ' + c.last_name
'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''
        select s.first_name + ' ' + s.last_name as full_name, st.store_name, count(o.order_id) as total_orders_made
        from orders o join staffs s on o.staff_id = s.staff_id
                      join stores st on o.store_id = st.store_id
        group by s.staff_id, s.first_name + ' ' + s.last_name, st.store_id, st.store_name
'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''
        with customer_stats as (
        select c.customer_id, c.first_name + ' ' + c.last_name as full_name, count(o.order_id) as total_orders,
               datediff(day, max(o.order_date), '2017-12-31') as days_since_last_order
        from orders o join customers c on o.customer_id = c.customer_id
        group by c.customer_id, c.first_name + ' ' + c.last_name
        )

        select customer_id, full_name,
            case 
                when total_orders > 1 then 'Repeat Customer'
                else 'New Customer'
            end as customer_type,

            case 
                when days_since_last_order < 60 then 'Active Customer'
                else 'Inactive Customer'
            end as customer_status
        from customer_stats
'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''
        create view rejected_orders_tracking as 
        select p.product_name, oi.quantity, o.order_date, o.required_date, 
               case 
                   when o.order_status = '1' then 'Pending'
                   when o.order_status = '2' then 'Processing'
                   when o.order_status = '3' then 'Rejected'
                   when o.order_status = '4' then 'Completed'
                end as order_status   
        from orders o join order_items oi on o.order_id = oi.order_id
                      join products p on oi.product_id = p.product_id
        where o.order_status = '3'
        group by p.product_name, oi.quantity, o.order_date, o.required_date,
                 case 
                   when o.order_status = '1' then 'Pending'
                   when o.order_status = '2' then 'Processing'
                   when o.order_status = '3' then 'Rejected'
                   when o.order_status = '4' then 'Completed'
                end
'''
# cursor.execute(query)
# connection.commit()
```

```python
df = pd.read_sql('select * from rejected_orders_tracking', connection)
df
```

```python
query = '''
        create view customer_orders_tracking as 
        select c.first_name + ' ' + c.last_name as full_name, count(o.order_id) as total_orders,
               min(o.order_date) as first_order_date, max(o.order_date) as last_order_date,
               sum(oi.quantity * oi.list_price * (1 - oi.discount)) as total_spent
        from orders o join customers c on o.customer_id = c.customer_id
                      join order_items oi on o.order_id = oi.order_id
        group by c.customer_id, c.first_name + ' ' + c.last_name
'''
# cursor.execute(query)
# connection.commit()
```

```python
df = pd.read_sql('select * from customer_orders_tracking', connection)
df
```

```python
query = '''
        create view stock_tracking as 
        select p.product_name, c.category_name, p.model_year, st.quantity, s.store_name, s.city 
        from products p join stocks st on p.product_id = st.product_id
                        join categories c on p.category_id = c.category_id
                        join brands b on p.brand_id = b.brand_id
                        join stores s on st.store_id = s.store_id
'''
# cursor.execute(query)
# connection.commit()
```

```python
query = '''
        select c.first_name + ' ' + c.last_name as full_name,
              c.city, count(o.order_id) as total_orders,
              rank() over (partition by c.city order by count(o.order_id) desc) as city_rank
        from orders o join order_items oi on o.order_id = oi.order_id
                     join customers c on o.customer_id = c.customer_id
        group by c.first_name + ' ' + c.last_name, c.city
'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''
        create view ranking_customers_by_orders_in_each_city as 
        select c.first_name + ' ' + c.last_name as full_name,
              c.city, count(o.order_id) as total_orders,
              rank() over (partition by c.city order by count(o.order_id) desc) as city_rank
        from orders o join order_items oi on o.order_id = oi.order_id
                     join customers c on o.customer_id = c.customer_id
        group by c.first_name + ' ' + c.last_name, c.city
'''
cursor.execute(query)
connection.commit()
```

```python
df = pd.read_sql('select * from ranking_customers_by_orders_in_each_city', connection)
df
```

```python
query = '''
        select p.product_name, c.category_name, sum(st.quantity) as total_stocks, 
               rank() over (partition by c.category_name order by sum(st.quantity) desc) as stock_rank_by_category
        from products p join categories c on p.category_id = c.category_id
                        join stocks st on p.product_id = st.product_id
        group by p.product_name, c.category_name
'''
df = pd.read_sql(query, connection)
df
```

```python
query = '''
        select p.product_name, s.store_name, sum(st.quantity) as total_stocks, 
               rank() over (partition by s.store_name order by sum(st.quantity) desc) as stock_rank_by_store
        from products p join stocks st on p.product_id = st.product_id
                        join stores s on st.store_id = s.store_id
        group by p.product_name, s.store_name
'''
df = pd.read_sql(query, connection)
df
```
